# **1. Abstract**
Dentro del ámbito de la ciencia de datos, comprender y mejorar la satisfacción de los pasajeros de aerolíneas es un objetivo fundamental. El objetivo principal de este estudio es construir un modelo predictivo que categorice con precisión a los pasajeros en dos niveles distintos de satisfacción con la aerolínea: "Satisfecho" y "Neutral o Insatisfecho".

Podemos plantearemos preguntas como: ¿Los pasajeros más jóvenes tienden a estar más satisfechos? ¿Existe una diferencia significativa en la satisfacción entre géneros? ¿El tipo de clase y la clasificación de servicios afectan la satisfacción del cliente? ¿Los retrasos en la salida y llegada impactan en la satisfacción del pasajero?

**Contexto Comercial:**
En la industria de la aviación, la satisfacción del cliente es una piedra angular para el éxito de cualquier aerolínea. Los pasajeros satisfechos no solo son más propensos a ser leales, sino que también pueden influir en la elección de aerolíneas por parte de otros viajeros. En este contexto, nuestro proyecto se presenta como una oportunidad estratégica para las aerolíneas para optimizar sus servicios y, en última instancia, aumentar la satisfacción del cliente y su cuota de mercado.

**Problema Comercial:**
El problema central al que se enfrentan las aerolíneas es cómo mejorar la satisfacción del cliente y mantener una ventaja competitiva en un mercado altamente competitivo. Para abordar este desafío, es crucial comprender qué factores influyen en la satisfacción del pasajero y cómo pueden ajustarse los servicios y operaciones en consecuencia. Nuestro proyecto se propone resolver este problema planteando preguntas fundamentales y explorando hipótesis para proporcionar información sobre los factores que impactan en la satisfacción del cliente.

Al comprender mejor estos factores, las aerolíneas podrán tomar decisiones más precisas y orientadas hacia la satisfacción del cliente, lo que se traducirá en una experiencia de viaje mejorada y una ventaja competitiva sostenible en la industria de la aviación.

**Hipótesis**

1) **Factor de Edad:** Se plantea la hipótesis de que los pasajeros más jóvenes podrían exhibir niveles más altos de satisfacción en comparación con los pasajeros mayores. Esto se debe a la suposición de que los individuos más jóvenes podrían adaptarse mejor a las comodidades y servicios modernos.

2) **Influencia del Género:** Esta hipótesis explora si existe una disparidad en la satisfacción de los pasajeros basada en el género. Se conjetura que el género podría desempeñar un papel en las preferencias por ciertos servicios y en la satisfacción general.

3) **Tipo de Clase:** La hipótesis sugiere que el tipo de vuelo, ya sea Eco o Bussisnes, podría influir en la satisfacción del pasajero. Se postula que los pasajeros en vuelos clase Bussisnes podrían tener expectativas más altas debido a la tarifa que se paga del boleto.

4) **Clasificación de los Servicios:** La hipótesis considera que la percepción de los pasajeros sobre diferentes categorías de servicios (por ejemplo, entretenimiento a bordo, calidad de la comida, comportamiento de la tripulación) está vinculada a la satisfacción general. Anticipamos que las calificaciones más altas en estas categorías de servicios se correlacionarán positivamente con la satisfacción.

5) **Retrasos en la Salida y Llegada:** Se plantea la hipótesis de que los pasajeros que experimentan mínimos retrasos en la salida y llegada tendrán una mayor probabilidad de estar satisfechos. Los retrasos suelen ser una fuente significativa de frustración, y una reducción de los mismos podría contribuir a una percepción más positiva de la aerolínea.

## 3.2. Verificación de datos no relevantes

```python
train.describe()
```

Vemos aquí que hay columnas que tienen valor 0 y eso no es posible (viendo el valor de minimo).

En las columnas que se rankeaba del 1 al 5 las caraterísitcas del vuelo, que haya un voto 0 significa que no hay votado.

Se aclara que si bien también hay valores de mínimo en Delay (ultimas dos columnas), esto sí es posible ya que puede no existir demora.

Si bien una de las posibilidades es reemplazar dichos valores por la media, ésta nos dará un valor con decimal y no un valor entero como son los votos, por lo que no podremos utilizar ésta técnica. Debemos proceder a eliminar dichas filas para eliminar éstos registros.

```python
# Creamos un dict seleccionando las columnas a actualizar y calculamos la media de las columnas a actualizar
columnas_a_verificar = ["Inflight wifi service", "Departure/Arrival time convenient", "Ease of Online booking", "Gate location", "Food and drink", "Online boarding", "Seat comfort", "Inflight entertainment", "On-board service", "Leg room service", "Baggage handling", "Checkin service", "Inflight service", "Cleanliness"]
```

```python
# Creamos una máscara booleana que identifique las filas con al menos un valor 0 en las columnas de interés.
mask = (train[columnas_a_verificar] == 0).any(axis=1)

# Utilizamos la máscara para filtrar las filas que cumplen con la condición.
filas_con_valor_cero = train[mask]

# Crea un nuevo DataFrame que contiene solo las filas que no cumplen con la condición.
train=train[~mask]

print("Cantidad de filas eliminadas:", len(filas_con_valor_cero))
print("Cantidad de filas restantes:", len(train))
```

## 3.4. Verificación de outliers

```python
ax = sns.boxplot(x="satisfaction", y="Age", data=train)
```

```python
ax = sns.boxplot(x="satisfaction", y="Flight Distance", data=train)
```

```python
ax = sns.boxplot(x="satisfaction", y="Departure Delay in Minutes", data=train)
```

```python
# Columnas con variables numericas a identificar outliers
columnas_numericas = ["Age", "Flight Distance", "Departure Delay in Minutes"]

# Detección de outliers para cada columna numérica
outliers = {}
for col in columnas_numericas:
    Q1 = train[col].quantile(0.25)
    Q3 = train[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    outliers[col] = train[(train[col] < lower_limit) | (train[col] > upper_limit)]

# Mostrar los outliers para cada columna numérica
for col, df_outliers in outliers.items():
    print(f"Outliers en la columna '{col}':")
    print(df_outliers)
```

Si bien se detectan que existen outiliers en las columnas "Age", "Flight Distance", "Departure Delay in Minutes", pueden ser datos relevantes para el modelo, por lo que se decide NO ELIMINARLOS.

# **8. Análisis Bias-variance tradeoff**

El análisis del bias-variance tradeoff es fundamental en la selección de modelos de aprendizaje automático. Aquí hay un ejemplo de cómo puedes probar diferentes modelos de clasificación en tu conjunto de datos y seleccionar el mejor en función del equilibrio entre sesgo y varianza. Usaré un conjunto de modelos comunes para este ejemplo.

En este código, utilizamos cross_val_score de scikit-learn para evaluar cada modelo con validación cruzada de 5 folds. Estamos utilizando la métrica de precisión (scoring='accuracy')

```python
from sklearn.model_selection import cross_val_score

# Resultados
results = {}

# Diccionario para almacenar los modelos y los tiempos de ejecución
models = {
    'RandomForest': RandomForestClassifier(max_depth=2, random_state=0),
    'AdaBoost': AdaBoostClassifier(random_state=0),
    'GradientBoosting': GradientBoostingClassifier(random_state=0),
    'GaussianNB': GaussianNB(),
    'KNeighbors': KNeighborsClassifier(),
    'XGBClassifier': XGBClassifier(),
    'LogisticRegression': LogisticRegression(random_state=0),
}

# Iterar sobre los modelos
for model_name, model in models.items():
    # Calcular cross_val_score con 5 folds
    scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')

    # Almacenar resultados
    results[model_name] = scores

# Imprimir resultados
for model_name, scores in results.items():
    print(f'{model_name}: Accuracy - Media: {np.mean(scores)}, Desviación Estándar: {np.std(scores)}')
```

# **9. Evaluación de las curvas ROC y tiempos**

Para finalizar, usemos ésto para entrenar disintos modelos y evaluar qué tal bien funciona, a través de la curva ROC!

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
import time

# Diccionario para almacenar los modelos y los tiempos de ejecución
models = {
    'RandomForest': RandomForestClassifier(max_depth=2, random_state=0),
    'AdaBoost': AdaBoostClassifier(random_state=0),
    'GradientBoosting': GradientBoostingClassifier(random_state=0),
    'GaussianNB': GaussianNB(),
    'KNeighbors': KNeighborsClassifier(),
    'XGBClassifier': XGBClassifier(),
    'LogisticRegression': LogisticRegression(random_state=0),
    'DecisionTree': DecisionTreeClassifier(random_state=0),
}

# Diccionario para almacenar los resultados y los tiempos de ejecución
results = {}
execution_times = []

# Iterar sobre los modelos
for model_name, model in models.items():
    # Iniciar el temporizador
    start_time = time.time()

    # Entrenar el modelo
    model.fit(X_train, y_train)

    # Realizar predicciones
    y_pred = model.predict(X_test)

    # Detener el temporizador
    end_time = time.time()

    # Calcular el tiempo de ejecución
    execution_time = end_time - start_time

    # Almacenar los resultados y el tiempo de ejecución
    results[model_name] = {'y_pred': y_pred, 'execution_time': execution_time}
    execution_times.append(execution_time)

    # Imprimir el tiempo de ejecución
    print(f"{model_name}: Tiempo de ejecución = {execution_time:.2f} segundos")

# Ahora `execution_times` contiene los tiempos de ejecución para cada modelo
```

```python
# Crear un diccionario para almacenar los tiempos de ejecución
tt = {}

# Iterar sobre los modelos y los tiempos de ejecución
for model_name, execution_time in zip(models.keys(), execution_times):
    tt[model_name] = execution_time

# Ahora `tt` contiene los tiempos de ejecución para cada modelo
print(tt)
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score, f1_score, recall_score, confusion_matrix

# Creación de una función para imprimir las métricas
def print_metrics(model_name, y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    roc_auc = roc_auc_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    confusion = confusion_matrix(y_true, y_pred)

    print(f"{model_name}:")
    print("Accuracy:", accuracy)
    print("ROC AUC:", roc_auc)
    print("F1 Score:", f1)
    print("Recall:", recall)
    print("Confusion Matrix:")
    print(confusion)
    print("\n")

# Entrenar y evaluar RandomForestClassifier
print_metrics("RandomForestClassifier", y_test, rf_pred)

# Entrenar y evaluar AdaBoostClassifier
print_metrics("AdaBoostClassifier", y_test, adb_pred)

# Entrenar y evaluar GradientBoostingClassifier
print_metrics("GradientBoostingClassifier", y_test, gdb_pred)

# Entrenar y evaluar GaussianNB
print_metrics("GaussianNB", y_test, nbc_pred)

# Entrenar y evaluar KNeighborsClassifier
print_metrics("KNeighborsClassifier", y_test, knn_pred)

# Entrenar y evaluar XGBClassifier
print_metrics("XGBClassifier", y_test, xgb_pred)

# Entrenar y evaluar LogisticRegression
print_metrics("LogisticRegression", y_test, logr_pred)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Crear una figura
plt.figure(figsize=(10, 8))

# Iterar sobre los modelos y calcular las curvas ROC-AUC
for model_name, model in zip(['RandomForest', 'AdaBoost', 'GradientBoosting', 'GaussianNB', 'KNeighbors', 'XGBClassifier', 'LogisticRegression'], [RandomForestClassifier(max_depth=2, random_state=0), AdaBoostClassifier(random_state=0), GradientBoostingClassifier(random_state=0), GaussianNB(), KNeighborsClassifier(), XGBClassifier(), LogisticRegression(random_state=0)]):
    # Ajustar el modelo
    model.fit(X_train, y_train)

    # Obtener probabilidades predichas
    y_prob = model.predict_proba(X_test)[:, 1]

    # Calcular la curva ROC
    fpr, tpr, _ = roc_curve(y_test, y_prob)

    # Calcular el área bajo la curva ROC (AUC)
    roc_auc = auc(fpr, tpr)

    # Graficar la curva ROC
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})')

# Configurar la gráfica
plt.plot([0, 1], [0, 1], 'k--', label='Random')
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curvas ROC-AUC de Modelos')
plt.legend(loc='lower right')

# Mostrar la gráfica
plt.show()
```

### **10.3.2. Stratified KFold**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
import numpy as np

# Crear el modelo RandomForestClassifier
model = RandomForestClassifier(random_state=42, n_estimators=30, criterion="gini", max_depth=4)

# Crear el procedimiento Stratified K-Fold
skf = StratifiedKFold(n_splits=5, random_state=42, shuffle=True)

# Lista para guardar accuracy de cada fold
fold_accuracies = []

# Iterar sobre los folds
for fold, (train_index, test_index) in enumerate(skf.split(X_train, y_train), 1):
    # Obtener conjuntos de entrenamiento y prueba
    X_train_fold, X_test_fold = X_train.iloc[train_index], X_train.iloc[test_index]
    y_train_fold, y_test_fold = y_train.iloc[train_index], y_train.iloc[test_index]

    # Ajustar el modelo
    model.fit(X_train_fold, y_train_fold)

    # Predecir en el conjunto de prueba
    y_pred_fold = model.predict(X_test_fold)

    # Calcular el accuracy y guardarlo en la lista
    accuracy_fold = accuracy_score(y_test_fold, y_pred_fold)
    fold_accuracies.append(accuracy_fold)

    print(f'Fold {fold}: Accuracy = {accuracy_fold}')

# Calcular y mostrar el accuracy promedio
average_accuracy = np.mean(fold_accuracies)
print(f'Accuracy promedio de los {skf.get_n_splits(X_train, y_train)} folds: {average_accuracy}')
```
